## Part 2: Image Augmentation with Albumentations

### 2.1 Creating Sample Images

```python
# Create a sample image (movie poster placeholder)
def create_sample_poster(width=400, height=600):
    """
    Create a synthetic movie poster image for demonstration.
    """
    # Create gradient background
    img = np.zeros((height, width, 3), dtype=np.uint8)
    
    # Create gradient from dark blue to black
    for i in range(height):
        ratio = i / height
        img[i, :, 0] = int(20 * (1 - ratio))   # Blue
        img[i, :, 1] = int(10 * (1 - ratio))   # Green
        img[i, :, 2] = int(40 * (1 - ratio))   # Red (in BGR)
    
    # Add some shapes to represent movie poster elements
    # Title area (rectangle)
    cv2.rectangle(img, (50, 450), (350, 550), (200, 180, 100), -1)
    
    # Character silhouette (ellipse)
    cv2.ellipse(img, (200, 250), (80, 150), 0, 0, 360, (100, 80, 60), -1)
    
    # Star rating
    for i in range(5):
        x = 80 + i * 60
        cv2.circle(img, (x, 580), 15, (0, 200, 255), -1)
    
    return img

# Create and display sample image
sample_image = create_sample_poster()

plt.figure(figsize=(6, 9))
plt.imshow(cv2.cvtColor(sample_image, cv2.COLOR_BGR2RGB))
plt.title("Sample Movie Poster")
plt.axis('off')
plt.show()

print(f"Image shape: {sample_image.shape}")
```

### Question 2.4: Implement MixUp

```python
# TODO: Implement MixUp augmentation
# MixUp blends two images and their labels

import numpy as np

def mixup(image1, image2, label1, label2, alpha=0.4):
    """
    Perform MixUp augmentation.

    Args:
        image1, image2: Two images (HxWxC numpy arrays)
        label1, label2: Their corresponding labels (one-hot or probabilities)
        alpha: Parameter for Beta distribution

    Returns:
        mixed_image: Blended image
        mixed_label: Blended label
    """
    # Sample lambda from Beta distribution
    lam = np.random.beta(alpha, alpha)

    # Blend images
    mixed_image = (lam * image1.astype(np.float32) +
                   (1 - lam) * image2.astype(np.float32)).astype(np.uint8)

    # Blend labels (soft labels)
    mixed_label = lam * np.array(label1, dtype=np.float32) +                   (1 - lam) * np.array(label2, dtype=np.float32)

    return mixed_image, mixed_label

# Create a second sample image
import cv2

sample_image2 = create_sample_poster()
# Modify it slightly (different colour tint)
sample_image2 = cv2.addWeighted(sample_image2, 0.7,
                                 np.full_like(sample_image2, [40, 80, 20]), 0.3, 0)

label1 = [1, 0, 0, 0, 0]   # e.g. Action (one-hot, 5 classes)
label2 = [0, 0, 1, 0, 0]   # e.g. Drama

mixed_img, mixed_lbl = mixup(sample_image, sample_image2, label1, label2, alpha=0.4)

fig, axes = plt.subplots(1, 3, figsize=(14, 5))
for ax, img, title in zip(axes, [sample_image, sample_image2, mixed_img],
                           ["Image 1 (Action)", "Image 2 (Drama)", "MixUp Blend"]):
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); ax.set_title(title); ax.axis('off')
plt.suptitle(f"MixUp Augmentation  |  mixed_label = {mixed_lbl.round(2)}")
plt.tight_layout(); plt.show()
print(f"Mixed label: {mixed_lbl}")
```

### Question 2.5: Augmentation with Bounding Boxes

```python
# Augmentation that preserves bounding box annotations
bbox_transform = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.Rotate(limit=15, p=0.5),
    A.RandomBrightnessContrast(p=0.3),
], bbox_params=A.BboxParams(format='pascal_voc', label_fields=['labels']))

# Sample bounding boxes [x_min, y_min, x_max, y_max]
bboxes = [
    [50, 450, 350, 550],   # Title area
    [120, 100, 280, 400],  # Character
]
labels = ['title', 'character']

# Apply transform
transformed = bbox_transform(
    image=sample_image,
    bboxes=bboxes,
    labels=labels
)

# Draw bounding boxes
def draw_bboxes(image, bboxes, labels):
    img = image.copy()
    colors = {'title': (0, 255, 0), 'character': (255, 0, 0)}
    for bbox, label in zip(bboxes, labels):
        x1, y1, x2, y2 = [int(c) for c in bbox]
        cv2.rectangle(img, (x1, y1), (x2, y2), colors.get(label, (0, 0, 255)), 2)
        cv2.putText(img, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, colors.get(label, (0, 0, 255)), 2)
    return img

# Display
fig, axes = plt.subplots(1, 2, figsize=(12, 8))

orig_with_boxes = draw_bboxes(sample_image, bboxes, labels)
axes[0].imshow(cv2.cvtColor(orig_with_boxes, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original with Bboxes")
axes[0].axis('off')

aug_with_boxes = draw_bboxes(transformed['image'], transformed['bboxes'], transformed['labels'])
axes[1].imshow(cv2.cvtColor(aug_with_boxes, cv2.COLOR_BGR2RGB))
axes[1].set_title("Augmented with Bboxes")
axes[1].axis('off')

plt.show()
```

### Question 3.5: Create a Complete Text Augmentation Pipeline

```python
# TODO: Create a pipeline that applies multiple augmentations
# Use naf.Sequential or naf.Sometimes for probabilistic application

import nlpaug.flow as naf
import nlpaug.augmenter.word as naw
import nlpaug.augmenter.char as nac

def create_text_augmentation_pipeline():
    """
    Create a text augmentation pipeline for movie review sentiment.

    Requirements:
    - Should preserve sentiment (don't replace sentiment words)
    - Should maintain readability
    - Should be diverse
    """
    pipeline = naf.Sequential([
        # 50% chance: synonym replacement (preserves meaning)
        naf.Sometimes([
            naw.SynonymAug(aug_src='wordnet', aug_max=2),
        ], pipeline_p=0.5),

        # 30% chance: mild random deletion (simulates noisy user input)
        naf.Sometimes([
            naw.RandomWordAug(action='delete', aug_p=0.1),
        ], pipeline_p=0.3),

        # 20% chance: keyboard typo (simulates real review typos)
        naf.Sometimes([
            nac.KeyboardAug(aug_char_max=1, aug_word_max=1),
        ], pipeline_p=0.2),
    ])
    return pipeline

# Test your pipeline
pipeline = create_text_augmentation_pipeline()
review = "This movie was absolutely fantastic! A must-watch."
try:
    augmented_texts = [pipeline.augment(review)[0] for _ in range(3)]
    print(f"Original: {review}\n")
    for i, aug in enumerate(augmented_texts, 1):
        print(f"Augmented {i}: {aug}")
except Exception as e:
    print(f"Pipeline test: {e}")
    print("(Pipeline created — run with installed nlpaug for full test)")
```

## Part 4: Audio Augmentation

### 4.1 Creating Sample Audio

```python
import numpy as np

# Create a synthetic audio signal (sine wave + noise)
def create_sample_audio(duration=2.0, sr=16000):
    """
    Create a sample audio signal.
    
    Args:
        duration: Duration in seconds
        sr: Sample rate
    
    Returns:
        audio: Audio signal as numpy array
    """
    t = np.linspace(0, duration, int(sr * duration), endpoint=False)
    
    # Create a combination of frequencies (like a chord)
    audio = (
        0.3 * np.sin(2 * np.pi * 440 * t) +   # A4 note
        0.2 * np.sin(2 * np.pi * 554 * t) +   # C#5 note
        0.2 * np.sin(2 * np.pi * 659 * t) +   # E5 note
        0.1 * np.random.randn(len(t))          # Some noise
    )
    
    # Normalize
    audio = audio / np.max(np.abs(audio))
    
    return audio.astype(np.float32)

sample_audio = create_sample_audio()
sr = 16000

print(f"Audio shape: {sample_audio.shape}")
print(f"Duration: {len(sample_audio) / sr:.2f} seconds")

# Plot waveform
plt.figure(figsize=(12, 3))
plt.plot(np.linspace(0, len(sample_audio)/sr, len(sample_audio)), sample_audio)
plt.title("Sample Audio Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

### Question 4.1: Basic Audio Augmentations

```python
# Manual audio augmentations

def add_noise(audio, noise_level=0.01):
    """Add Gaussian noise to audio."""
    noise = np.random.randn(len(audio)) * noise_level
    return audio + noise

def time_shift(audio, shift_range=0.2):
    """Shift audio in time."""
    shift = int(len(audio) * shift_range * np.random.uniform(-1, 1))
    return np.roll(audio, shift)

def change_volume(audio, volume_range=(0.5, 1.5)):
    """Change audio volume."""
    volume = np.random.uniform(*volume_range)
    return audio * volume

def pitch_shift_simple(audio, shift_factor=0.1):
    """Simple pitch shift by resampling."""
    from scipy.signal import resample
    factor = 1 + np.random.uniform(-shift_factor, shift_factor)
    new_length = int(len(audio) / factor)
    shifted = resample(audio, new_length)
    # Pad or trim to original length
    if len(shifted) > len(audio):
        return shifted[:len(audio)]
    else:
        return np.pad(shifted, (0, len(audio) - len(shifted)))

# Apply augmentations
augmentations = {
    'Original': sample_audio,
    'Added Noise': add_noise(sample_audio, 0.05),
    'Time Shifted': time_shift(sample_audio, 0.1),
    'Volume Changed': change_volume(sample_audio, (0.5, 1.5)),
}

# Plot
fig, axes = plt.subplots(4, 1, figsize=(12, 8))
for idx, (name, audio) in enumerate(augmentations.items()):
    axes[idx].plot(audio)
    axes[idx].set_title(name)
    axes[idx].set_xlim(0, len(audio))
plt.tight_layout()
plt.show()
```

### Question 4.2: Implement SpecAugment

```python
# TODO: Implement SpecAugment (frequency and time masking on spectrograms)

from scipy.signal import spectrogram as scipy_spectrogram
import numpy as np
import matplotlib.pyplot as plt

def spec_augment(audio, sr, freq_mask_param=10, time_mask_param=20):
    """
    Apply SpecAugment to audio.

    Args:
        audio: Audio signal (1-D numpy array)
        sr: Sample rate
        freq_mask_param: Maximum frequency mask width (in spectrogram bins)
        time_mask_param: Maximum time mask width (in frames)

    Returns:
        Augmented spectrogram (2-D numpy array)
    """
    # Step 1: Compute spectrogram
    freqs, times, Sxx = scipy_spectrogram(audio, fs=sr, nperseg=256)
    # Convert to log-power for stability
    Sxx_log = np.log1p(np.abs(Sxx))

    n_freqs, n_frames = Sxx_log.shape
    augmented = Sxx_log.copy()

    # Step 2: Frequency masking — zero out a band of frequency bins
    f_mask_width = np.random.randint(0, freq_mask_param + 1)
    f0 = np.random.randint(0, max(1, n_freqs - f_mask_width))
    augmented[f0: f0 + f_mask_width, :] = 0.0

    # Step 3: Time masking — zero out a contiguous block of frames
    t_mask_width = np.random.randint(0, time_mask_param + 1)
    t0 = np.random.randint(0, max(1, n_frames - t_mask_width))
    augmented[:, t0: t0 + t_mask_width] = 0.0

    return augmented, freqs, times

# Test
audio, sr = create_sample_audio()
aug_spec, freqs, times = spec_augment(audio, sr, freq_mask_param=10, time_mask_param=20)
_, _, orig_spec_raw = scipy_spectrogram(audio, fs=sr, nperseg=256)
orig_spec = np.log1p(np.abs(orig_spec_raw))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].pcolormesh(times, freqs, orig_spec, shading='gouraud')
axes[0].set_title("Original Spectrogram"); axes[0].set_ylabel("Frequency (Hz)"); axes[0].set_xlabel("Time (s)")
axes[1].pcolormesh(times, freqs, aug_spec, shading='gouraud')
axes[1].set_title("SpecAugmented"); axes[1].set_xlabel("Time (s)")
plt.tight_layout(); plt.show()
print("SpecAugment applied: frequency and time masking demonstrated.")
```

## Part 5: Measuring Augmentation Impact

### 5.1 Setting Up a Classification Task

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load a small subset of 20 newsgroups
categories = ['rec.sport.baseball', 'rec.sport.hockey']
newsgroups = fetch_20newsgroups(subset='all', categories=categories, 
                                 remove=('headers', 'footers', 'quotes'))

# Limit to smaller dataset for demonstration
texts = newsgroups.data[:500]
labels = newsgroups.target[:500]

# Split
X_train, X_test, y_train, y_test = train_test_split(
    texts, labels, test_size=0.3, random_state=42
)

print(f"Training samples: {len(X_train)}")
print(f"Test samples: {len(X_test)}")
```

### Question 5.1 (Solved): Baseline Without Augmentation

```python
# SOLVED: Train baseline model without augmentation

# Vectorize
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Train
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)

# Evaluate
y_pred = model.predict(X_test_vec)
baseline_accuracy = accuracy_score(y_test, y_pred)

print(f"Baseline Accuracy (no augmentation): {baseline_accuracy:.2%}")
```

### Question 5.2: Train with Augmentation

```python
# TODO: Augment training data and measure improvement
# 1. Augment each training sample 2-3 times
# 2. Combine with original data
# 3. Train model
# 4. Compare accuracy

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

delete_aug = naw.RandomWordAug(action='delete', aug_p=0.15)
synonym_aug_s = naw.SynonymAug(aug_src='wordnet', aug_max=2)

augmented_texts  = list(X_train)
augmented_labels = list(y_train)

print("Augmenting training data...")
for text, label in zip(X_train, y_train):
    try:
        aug1 = delete_aug.augment(text)[0]
        augmented_texts.append(aug1); augmented_labels.append(label)
        aug2 = synonym_aug_s.augment(text)[0]
        augmented_texts.append(aug2); augmented_labels.append(label)
    except Exception:
        pass   # skip if augmentation fails for this sample

print(f"Original training samples : {len(X_train)}")
print(f"After augmentation        : {len(augmented_texts)}")

# Vectorize
X_aug_vec = vectorizer.transform(augmented_texts)
X_test_vec2 = vectorizer.transform(X_test)

# Train augmented model
model_aug = LogisticRegression(max_iter=1000)
model_aug.fit(X_aug_vec, augmented_labels)
aug_accuracy = accuracy_score(y_test, model_aug.predict(X_test_vec2))

print(f"\nBaseline accuracy (no augmentation): {baseline_accuracy:.2%}")
print(f"Augmented accuracy                  : {aug_accuracy:.2%}")
print(f"Improvement                         : {aug_accuracy - baseline_accuracy:+.2%}")
```

### Question 6.2: Create an Augmentation Checklist

```python
# TODO: Write a checklist for designing augmentation pipelines

augmentation_checklist = """
# Augmentation Design Checklist

## Before Choosing Augmentations:
[ ] What is the task? (classification, detection, segmentation, generation, etc.)
[ ] What modality? (image, text, audio, video, tabular)
[ ] What domain? (natural images, medical scans, movie posters, social media text, etc.)
[ ] What is the dataset size? (small → more augmentation; large → less critical)
[ ] What is the class imbalance? (minority classes may need more augmentation)

## Validity Questions — ask for EACH augmentation:
[ ] Does this augmentation PRESERVE the label?
     Example: horizontal flip OK for scene photos, NOT OK for text/digit recognition.
[ ] Does this augmentation change the semantics in an uncontrolled way?
     Example: synonym replacement should not flip sentiment words.
[ ] Is this augmentation realistic / in-distribution?
     Example: extreme rotations (>45°) on movie posters are unlikely in production.
[ ] Does this augmentation increase diversity meaningfully?
     Example: adding 10 brightness variants of the same image adds less value than
     applying 10 geometrically distinct transforms.

## Implementation:
[ ] Start conservative (low probability, low intensity) — verify quality first.
[ ] Use a fixed random seed for reproducibility during development.
[ ] Implement augmentation INSIDE the training loop (online) or as a pre-processing
     step (offline) — be explicit about which approach you choose.
[ ] For detection/segmentation tasks, ensure bounding boxes / masks are transformed
     consistently with the image.
[ ] Test augmentations visually on 20–50 samples before running full training.

## Testing:
[ ] Visual spot-check: augmented samples look plausible to a human.
[ ] Label-preservation check: a classifier should not be confused by augmented inputs.
[ ] Train-with / train-without comparison on a validation set.
[ ] Check that augmentation does not cause data leakage (e.g., don't augment test data).
[ ] Monitor training loss and validation accuracy for signs of over-augmentation
     (model underfits if augmentation is too aggressive).
"""

# Complete the checklist
print(augmentation_checklist)
```
